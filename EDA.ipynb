### FICO Average Distribution by Credit score

```python
fico_avg = (ds1["high_fico"] + ds1["low_fico"])/2.0
```

```python
ds1_w_ficoavg = ds1.copy()
ds1_w_ficoavg["fico_avg"] = fico_avg
```

```python
ds1_w_ficoavg
```

```python
grades = ds1_w_ficoavg["lc_credit_score"].unique()
grades
```

```python
ds1_w_ficoavg[ds1_w_ficoavg["lc_credit_score"]=="A"]
```

```python
ds_grouped = {}

for i,grade in enumerate(grades):
    print(i)
    ds_grouped_instance = ds1_w_ficoavg[ds1_w_ficoavg["lc_credit_score"]==grade]
    ds_grouped[grade] = ds_grouped_instance
```

```python
[len(ds_grouped[key]) for key in ds_grouped]
```

```python
ds_grouped["A"]
```

```python
fig = px.histogram(ds_grouped["D"], x="fico_avg",nbins=50, histnorm='', 
                   title='FICO Range Distribution Chart')

fig.update_layout(xaxis_title='FICO_Average', yaxis_title='FICO High - Low')
fig.show()
```

```python
[len(ds_grouped[key]) for key in ds_grouped]
```

```python
from plotly.subplots import make_subplots
import plotly.express as px

fig = make_subplots(
    rows=3, cols=3,
    subplot_titles=[f'Grade {g}' for g in ['A','B','C','D','E','F','G']]
)

fig.add_trace(px.histogram(ds_grouped["A"], x="fico_avg",histnorm="probability", nbins=50).data[0], row=1, col=1)
fig.add_trace(px.histogram(ds_grouped["B"], x="fico_avg", histnorm="probability", nbins=50).data[0], row=1, col=2)
fig.add_trace(px.histogram(ds_grouped["C"], x="fico_avg",  histnorm="probability", nbins=50).data[0], row=1, col=3)
fig.add_trace(px.histogram(ds_grouped["D"], x="fico_avg",  histnorm="probability", nbins=50).data[0], row=2, col=1)
fig.add_trace(px.histogram(ds_grouped["E"], x="fico_avg",  histnorm="probability", nbins=50).data[0], row=2, col=2)
fig.add_trace(px.histogram(ds_grouped["F"], x="fico_avg",  histnorm="probability", nbins=50).data[0], row=2, col=3)
fig.add_trace(px.histogram(ds_grouped["G"], x="fico_avg",  histnorm="probability", nbins=50).data[0], row=3, col=1)

fig.update_layout(height=800, width=1000, title_text="FICO Average Distribution per Grade")
fig.update_xaxes( dtick=25)
fig.show()
```

My conclusion is that FICO and Credit score are not directly related rather Credit score may use FICO as PART of its calculation

Finding: Customers with higher grades are more likely to have higher FICO averages.

### Another perspective of correlation: Distribution Comparision of FICO Averages By Grade

```python
fig = go.Figure()
colors = ["red","orange","yellow","green","blue","brown","black"]
for ind,key in enumerate(ds_grouped):
    data = ds_grouped[key]
    kde = gaussian_kde(data["fico_avg"])
    x_vals = np.linspace(200,900,100)
    y_vals = kde(x_vals) 

    fig.add_trace(go.Scatter(x=x_vals,y=y_vals,mode="lines",name=f"{key}",fill="tozeroy",line={"color":colors[ind]},opacity=0.0))
fig.update_layout(title="Normalized FICO Distribution (Grade A)", xaxis_title="FICO Avg", yaxis_title="Density")
fig.show()
```

This view confirms that 
 - The lower the grade (->G), the more Leptokurtic the graph
 - The higher the grade (A<-), the more Playkurtic the graph
